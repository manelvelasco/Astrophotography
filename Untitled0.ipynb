```python
from scipy import fftpack
import numpy as np
import pylab as py
from PIL import Image
import matplotlib.pyplot as plt
```

```python
img = np.asarray(Image.open('M99.jpg'))
py.imshow(img)
```

Fem la transformada de fourier...

```python
F1 = fftpack.fft2(img)
```

I ordenem els quadrants per tal que les baixes freqëncies apareguin al centre de la imatge

```python
F2 = fftpack.fftshift( F1 )
```

Calculem l'espectre de potencies en 2D

```python
psd2D = np.abs( F2 )**2
```

```python
fig,ax=f, ax = plt.subplots(1,2)
fig.set_size_inches(20,10)
ax[0].imshow( np.log10( img ))
ax[1].imshow( psd2D)
```
